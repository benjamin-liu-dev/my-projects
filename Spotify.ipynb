```python
# Replace with your Spotify app credentials
CLIENT_ID = 'your_client_id'
CLIENT_SECRET = 'your_client_secret'
REDIRECT_URI = 'http://localhost:8888/callback'

# Define the scope of access
SCOPE = 'user-top-read user-read-recently-played'

# Authenticate and get token
sp = spotipy.Spotify(auth_manager=SpotifyOAuth(
    client_id=CLIENT_ID,
    client_secret=CLIENT_SECRET,
    redirect_uri=REDIRECT_URI,
    scope=SCOPE
))

# Get top tracks
top_tracks = sp.current_user_top_tracks(limit=10, time_range='medium_term')
print("ğŸµ Your Top Tracks:")
for idx, item in enumerate(top_tracks['items']):
    print(f"{idx+1}. {item['name']} by {', '.join(artist['name'] for artist in item['artists'])}")

# Get recently played tracks
recent_tracks = sp.current_user_recently_played(limit=10)
print("\nğŸ•’ Recently Played Tracks:")
for idx, item in enumerate(recent_tracks['items']):
    track = item['track']
    print(f"{idx+1}. {track['name']} by {', '.join(artist['name'] for artist in track['artists'])}")
```
